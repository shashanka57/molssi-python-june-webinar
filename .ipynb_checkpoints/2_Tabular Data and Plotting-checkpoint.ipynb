```python
import numpy as np
import os
```

```python
help(np.genfromtxt)
```

```python
distance_file=os.path.join('data','distance_data_headers.csv')
distances= np.genfromtxt(fname=distance_file, delimiter=',',dtype='unicode')
print(distances)
```

```python
# dtype as unicode reads all the data as string. Necessity:genfromtxt is going to try to make all the data the same type.  Numbers (floats) is the default. Since we have multiple data types, we need to tell it unicode.
```

```python
headers = distances[0]
print(headers)
```

```python
#Slice distances to get the numerical values(everything except the header)
data=distances[1:]
```

```python
#To convert the data type from string
data=data.astype(np.float)
print(data)
```

```python
#To slice a 2D
#array_name[row index, column index]

print(data[0,1])
print(data[1,0])
```

```python
small_data=data[0:10,0:3] #rows first 10 rows and first 3 columns
print(small_data)
```

```python
#for avg. np.mean(data_set)
#avg of a column
thr4_atp=data[:,1]#variable name is the column name of the table we are using
avg_thr4_atp=np.mean(thr4_atp)
print(avg_thr4_atp)
```

```python
#avg each column
for i in range(1,len(headers)):
    avg_columns=np.mean(data[:,i])
    data_label=headers[i]
    print(F'{data_label} : {avg_columns:.3f}')
```

```python
#Plotting
import matplotlib.pyplot as plt

# matplotlib.inline is a jupyter notebook command that just says to show the plots in the notebook.  In old versions you had to have it.
#But it was so popular that they made it the default in new versions.
```

```python
plt.figure()
plt.plot(data[:,1])
```

```python
plt.figure()

sample_name=headers[1]

plt.xlabel('Simulation Frame')
plt.ylabel('Distance (A)')
plt.plot(data[:,1], label=sample_name)
plt.legend()

plt.savefig(F'{sample_name}.png',dpi=300) #dpi pixels per inch for higher resolution images
```

```python
plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (A)')
plt.plot(data[:,1], label=headers[1], color='green')
plt.plot(data[:,2], label=headers[2])
plt.legend()

plt.savefig('two_samples.png',dpi=300) #dpi pixels per inch for higher resolution images
```

```python
num_columns=len(headers)

for col in range(1,num_columns):
    plt.figure()
    plt.plot(data[:,col], label=headers[col])
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (A)')
    plt.legend()
    plt.savefig(F'{headers[col]}.png')

```

```python
# for having a common y axis range 
data_min=np.min(data[:,1:])
data_max=np.max(data[:,1:])
# then put another command
plt.ylim(data_min,data_max)
```

```python
#for every other row i.e. alternate rows 
print(data[:10:2,:])
```

```python
#for every alt row
plt.figure()
plt.plot(data[:10:2,0], data[:10:2,1])
```

```python
#every alt 100 row and line style which has to be after x and before label 
plt.figure() #'--','.'
plt.plot(data[::100,0],data[::100,1],'-o',label=headers[1])
plt.legend()
```
